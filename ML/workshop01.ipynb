```python
amazon[amazon['Courier Status'].isnull()]["Status"].unique()
```

```python
amazon['Courier Status'] = amazon.apply(lambda row: row['Status'].split('-')[0].strip()
                           if pd.isnull(row['Courier Status']) else row['Courier Status'],
                                        axis=1)
```

```python
print(amazon['Courier Status'])
```

```python
amazon['Date'] = pd.to_datetime(amazon['Date'])
```

```python
amazon['month'] = amazon['Date'].dt.to_period('M')
amazon['month'].unique()
```

```python
avg_price = amazon.groupby('Category')['Amount'].mean()
avg_price = avg_price.sort_values(ascending=False)
print("category prices data:")
print(avg_price)
```
